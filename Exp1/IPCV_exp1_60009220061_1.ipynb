```python
if img1.shape != img2.shape:
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
```

```python
add_img = cv2.add(img1,img2)
```

```python
def show_subplot(img1,img2,img3,title1,title2,title3):
    fig = plt.figure(figsize=(10,7))
    fig.add_subplot(3,3,1)
    plt.imshow(img1)
    plt.axis('off')
    plt.title(title1)
    fig.add_subplot(3,3,2)
    plt.imshow(img2)
    plt.axis('off')
    plt.title(title2)
    fig.add_subplot(3,3,3)
    plt.imshow(img3)
    plt.axis('off')
    plt.title(title3)
```

```python
show_subplot(img1,img2,add_img,"img1","img2","added image")
```

```python
sub_img = cv2.subtract(img1,img2)
show_subplot(img1,img2,sub_img,"img1","img2","subtracted image")
```

```python
#bitwise opeartions
bit_add = cv2.bitwise_and(img1,img2)
show_subplot(img1,img2,bit_add,"img1","img2","bitwise add image")
```

```python
# plt.imshow(bit_add)
```

```python
bit_or = cv2.bitwise_or(img1,img2)
show_subplot(img1,img2,bit_or,"img1","img2","bitwise or image")
```

```python
bit_not = cv2.bitwise_not(img1)
fig = plt.figure(figsize=(10,7))
fig.add_subplot(2,2,1)
plt.imshow(img1)
plt.axis('off')
plt.title("Image 1")
fig.add_subplot(2,2,2)
plt.imshow(img2)
plt.axis('off')
plt.title("Bitwise not image")
plt.imshow(bit_not)
```

```python
bit_xor = cv2.bitwise_xor(img1,img2)
show_subplot(img1,img2,bit_xor,"img1","img2","bitwise xor image")
```

```python
type(bit_xor)
```

```python
alpha = 0.5
beta = 0.5
gamma = 0

b = cv2.addWeighted(img1,alpha, img2, beta, gamma)
show_subplot(img1,img2,b,"img1","img2","Blended image")
```
